```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
```

Создаем датасет (с фантазией проблемы, поэтому он похож на тот, что в примере)

```python
# Создаём сэмпл
n_samples = 500

sn = np.random.choice(1000, n_samples, replace=False)
length = np.random.choice(120, n_samples) + np.random.choice(100, n_samples)
width = np.random.choice(80, n_samples) + np.random.choice(100, n_samples)
height = np.random.choice(200, n_samples) + np.random.choice(100, n_samples)
q_class = np.random.choice(3, n_samples) + 3


# vol = length * width * height
price = 10*length + 2*(width/height) + 5*q_class + 500 #линейная модель

data = pd.DataFrame({'sn': sn, 'length': length, 'width': width, 'height': height, 'q_class': q_class, 'price': price})
data
```

```python
# ?np.random.choice
```

```python
from sklearn.metrics import mean_absolute_error

X = data[['sn', 'length', 'width', 'height', 'q_class']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['sn', 'length', 'width', 'height', 'q_class']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

Отбросим переменную с наименьшим весом

```python
X = data[['length', 'width', 'height', 'q_class']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['length', 'width', 'height', 'q_class']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
# Создаем новый признак на основе трех размеров
data['vol'] = data['length'] * data['width'] * data['height']
data
```

```python
X = data[['vol', 'q_class']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['vol', 'q_class']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

Создадим еще признак максимально учитывающий все используемые

```python
data['koef'] = 10*data['length'] + 2*data['width']/data['height'] + 5*data['q_class']
data
```

```python
X = data[['koef']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['koef']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
